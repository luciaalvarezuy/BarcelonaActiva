Esta actividad se realiza en dos partes.
Primera parte, se realiza la siguiente consulta sobre la recién creada base de datos: Tienes que obtener, para cada persona, su nombre y el número máximo de roles distintos que ha tenido en una misma película. 

```python
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="[INGRESAR LA CONTRASEÑA]",
    database = "movies"
)

mycursor = mydb.cursor()


mycursor.execute("SELECT tb_person.person_id,\
                 tb_person.person_name,\
                 count(tb_movie_person.role_id)\
FROM ((`tb_movie_person`\
INNER JOIN `tb_person` ON tb_movie_person.person_id = tb_person.person_id)\
INNER JOIN `tb_role` ON tb_role.role_id = tb_movie_person.role_id)\
                  GROUP BY tb_movie_person.person_id\
                  ORDER BY tb_person.person_id ASC;")
myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```

Mostrar únicamente a aquellas personas que hayan asumido más de un rol en una misma película.

```python
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="[INGRESAR LA CONTRASEÑA]",
    database = "movies"
)

mycursor = mydb.cursor()


mycursor.execute("SELECT tb_person.person_id,\
                 tb_person.person_name,\
                 COUNT(tb_movie_person.role_id) \
FROM ((`tb_movie_person`\
INNER JOIN `tb_person` ON tb_movie_person.person_id = tb_person.person_id)\
INNER JOIN `tb_role` ON tb_role.role_id = tb_movie_person.role_id)\
                 GROUP BY tb_movie_person.person_id, tb_movie_person.movie_id\
                 HAVING COUNT(tb_movie_person.role_id)>1\
                  ORDER BY tb_person.person_id ASC;")
myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```
