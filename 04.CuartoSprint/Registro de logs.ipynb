## S04 T01: Transformación Registro Log con Regular expresiones

```python
import pandas as pd
import re
```

```python
#Cargar el dataframe
#Para vizualizar todo el contenido del documento
pd.set_option('display.max_colwidth', None)

df = pd.read_csv('C:/Users/Lucía Alvarez/OneDrive - psico.edu.uy/00 Barcelona activa/Web_access_log-akumenius.com.txt', 
                 delimiter = "\t", header=None)

df.columns =['Primera_variable']

df
```

```python
Nueva_variable = df["Primera_variable"].str.split('- -', expand=True)
Nueva_variable
```

```python
#Me interesa únicamente lo que aparece en la columna 0
df=df.assign(Variable_Creada=Nueva_variable[0])  
df
```

```python
#Separar la variable creada por el espacio que tiene entre los datos
new = df['Variable_Creada'].str.split(" ", n = 1, expand = True) 
  
df["Página WEB"]= new[0] 
  
df["IP"]= new[1] 
  
df 
```

```python
#Eliminar variable creada y que ahora no me sirve

df.drop(['Variable_Creada'], axis = 'columns', inplace=True)
df
```

```python
#Separar los datos que están en formato fecha y hora del dataframe
new = df['Primera_variable'].str.split(r"\[(.*?)\]", expand= True)
df["Fecha y hora"]= new[1] 
df
```

```python
# Se genera un nuevo objeto denominado "Objeto" para guardar los elementos que vamos a ir creando
Columnas_Nuevas=df["Primera_variable"].apply(lambda columna: re.findall("[^\"]*",columna))
Columnas_Nuevas
```

```python
# A continuación, creamos una nueva columna denominada "MensajeHTTP" que va a 
#guardar la informacióon que esstá en la segunda columna

df["MensajeHTTP"]=Columnas_Nuevas.map(lambda columna: columna[2])
df
```

```python
df["Numeros"]=Columnas_Nuevas.map(lambda columna: columna[4] if len(columna)>=5 else "")
```

```python
df["Pagina"]=Columnas_Nuevas.map(lambda columna: columna[6] if len(columna)>=7 else "")
df
```

```python
df["Navegador"]=Columnas_Nuevas.map(lambda columna: columna[10] if len(columna)>=11 else "")
df
```

```python
#Eliminar la primer variable
df_transformada= df.drop(['Primera_variable'], axis = 'columns', inplace=True)
df_transformada
```

```python
#Separar los datos que están en formato fecha y hora del dataframe
new2 = df['Fecha y hora'].str.split(r"[:]", expand= True)
df["Fecha"]= new2[0]
df


df['Fecha'].value_counts()
```
