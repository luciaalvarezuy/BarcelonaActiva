Crea una función que dado un Array de una dimensión, te haga un resumen estadístico básico de los datos. Si detecta que el array tiene más de una dimensión, debe mostrar un mensaje de error.

```python
import numpy as np

import statistics
from statistics import mean, mode, multimode, stdev

#Crear un array unidimensional
dataOneDimension = np.random.randint(10, size=(6)) 

#Crear un array con dos dimensiones
dataTwoDimensions = np.random.randint(10, size=(2, 6)) 


def ResumenDeEstadistico(data):
    if data.ndim == 1:
        print("La media del array es", statistics.mean(data), "su desviación estandar es de",
              round(float(stdev(data)), 1), "La mediana de la lista es", statistics.median(data),
              "con moda de", statistics.mode(data))
    else: 
        print("Error, el array tiene más de una dimensión")

#Testear la función con el array de una dimensión
ResumenDeEstadistico(dataOneDimension)

#Testear la función con el array de dos dimensiones
ResumenDeEstadistico(dataTwoDimensions)
```

```python
import math, statistics


def correlacion(x,y):
    #Primer paso: Calcular el promedio de las variables
    suma_x = sum(x)
    suma_y = sum(y)
    n = len(x)
    x_mean = statistics.mean(x)
    y_mean = statistics.mean(y)
    dividendo = 0.0
    divisor = 0.0
    Tercer_paso_x = 0.0
    Tercer_paso_y = 0.0
    #Segundo paso, calcular el dividendo (arriba)
    for i in range(n):
        dividendo += (x[i]-x_mean)*(y[i]-y_mean)
    #Tercer paso, calcular el divisor (abajo)
    for i in range(n):
        Tercer_paso_x += (x[i] - x_mean)*(x[i]-x_mean)
    for i in range(n):
        Tercer_paso_y += (y[i] - y_mean) * (y[i] - y_mean)
    divisor = math.sqrt(Tercer_paso_x*Tercer_paso_y)
    #Cuarto paso: realizar la división
    resultado = dividendo/divisor
    print("La correlación entre las variables es ", resultado)

# Evaluar si la función es correcta
x_simple = np.array([-2, -1, 0, 1, 2])
y_simple = np.array([4, 1, 3, 2, 0])
correlacion(x_simple, y_simple)


#Para comprobar el resultado de la función con la formula del programa
print(np.corrcoef(x_simple, y_simple))
```
