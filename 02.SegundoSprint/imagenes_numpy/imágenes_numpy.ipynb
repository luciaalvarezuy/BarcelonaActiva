```python
### Ejercicio 1
# Crea un np.array de una dimensión, que incluya al menos 8 números enteros, data type int64. 
import numpy as np
np_array = np.array([[1,2,3,4,5,6,7,8]], dtype="int64")

#Muestra la dimensión
print("Tenemos ", np_array.ndim , " dimensiones")

#La forma de la matriz.
print("La forma de la matriz es: ", np.shape(np_array))
```

```python
### Ejercicio 2
#De la matriz del ejercicio 1, calcula el valor medio de los valores introducidos 

elementos = 0
sumatoria = 0
for fila in np_array:
    for elemento in fila:
        sumatoria += elemento
        elementos += 1

promedio = sumatoria / elementos
print("El promedio de los valores introducidos en la matriz es ", promedio)

#y resta la media resultante de cada uno de los valores de la matriz.

matC = np_array - promedio
print("La matriz de los valores restando la media es: ", matC)
```

```python
###Ejercicio 3
#Crea una matriz bidimensional con una forma de 5 x 5. 

MatrizBidimensional = np.array([[11, 12, 13, 14, 15],
    [21, 22, 23, 24, 25],
    [31, 32, 33, 34, 35],
    [41, 42, 43, 44, 45],
    [51, 52, 53, 54, 55]])

#Extrae el valor máximo de la matriz

valorMaximo = np.amax(MatrizBidimensional)
print("El valor máximo del array es", valorMaximo)

#y los valores máximos de cada uno de sus ejes.

valores = []
for l in MatrizBidimensional:
    maximo = max(l)
    valores.append((maximo))

print("Los valores máximo de cada eje son: ", valores, "respectivamente")
```

```python
####Ejercicio 4
#Muéstrame con ejemplos de diferentes matrices, la regla fundamental de Broadcasting que dice:
#"las matrices se pueden transmitir / broadcast  si sus dimensiones coinciden o si una de las matrices tiene un tamaño de 1".

Matriz_a = np.array((10, 10, 10))
Matriz_b = np.array((1, 2, 3))
Matriz_c = np.array((1))

#Observar las matrices
#print(Matriz_a)
#print(Matriz_b)
#suma entre matrices con las mismas dimensiones
sumaDeMatrices1=Matriz_a + Matriz_b
print("Suma de matrices con dimensiones que coinciden: ", sumaDeMatrices1)

#Suma entre matriz con tamaño 1
sumaDeMatrices2=Matriz_a + Matriz_c
print("Suma de matrices con una dimensión con un tamaño de 1: ", sumaDeMatrices2)
```

```python
###Ejercicio 5
#Use la Indexación para extraer los valores de una columna y una fila de la matriz. 

Columna = np.array([MatrizBidimensional[0, 0], MatrizBidimensional[0, 1], 
      MatrizBidimensional[0, 2], MatrizBidimensional[0, 3], MatrizBidimensional[0, 4]])
                                
print('Los valores de la primer columna de la matriz bidimensional son: ', Columna)

Fila = np.array([MatrizBidimensional[0, 0], MatrizBidimensional[1, 0], 
      MatrizBidimensional[2, 0], MatrizBidimensional[3, 0], MatrizBidimensional[4, 0]])

print('Los valores de la primer fila de la matriz bidimensional son: ', Fila)

#Y suma sus valores.
sumaDeMatrices1=Columna + Fila
print("La matriz con la suma de los valores es:", sumaDeMatrices1)
```

```python
###Ejercicio 6
#Mask la matriz anterior, realice un cálculo booleano vectorizado, cogiendo cada elemento y comprobando si se divide uniformemente por cuatro.
#Esto devuelve una matriz de mask de la misma forma con los resultados elementales del cálculo.

#Importar paquete para enmascarar la matriz
import numpy.ma as ma

#Crear una matriz enmascarada, en este caso marcamos la cuarta entrada como inválida. 
sumaDeMatrices1Enmascarada = ma.masked_array(sumaDeMatrices1, mask=[0, 0, 0, 1, 0])
print (sumaDeMatrices1Enmascarada)

#realice un cálculo booleano vectorizado, cogiendo cada elemento y comprobando si se divide uniformemente por cuatro.
sumaDeMatrices1Enmascarada2 = sumaDeMatrices1Enmascarada % 4 ==0
print (sumaDeMatrices1Enmascarada2)
```

```python
###Ejercicio 7
#Utilice esta máscara para indexar en la matriz de números original. 
#Esto hace que la matriz pierda su forma original,  reduciéndola a una dimensión, pero todavía obtenga los datos que está buscando.

sumaDeMatrices1[~sumaDeMatrices1Enmascarada2]
```
