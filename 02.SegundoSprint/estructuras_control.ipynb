### Ejercicio 3
Crea un programa que te pregunte tu nombre, y te solicite un número. 
Si el número es 0, debería mostrar un mensaje de error. 
De lo contrario, debería mostrar el nombre repetido tantas veces como indique el número. 
Por ejemplo, "Joan Joan Joan".


```python
def programa():
    nombre = str(input('Escribe tu nombre: '))
    numeroARepetir = int(input('Escribe un número: '))
    if numeroARepetir == 0:
        print("Error")
    elif numeroARepetir <0:
        numeroARepetir = numeroARepetir * -1
        nombre = nombre + " "
        print( nombre  * numeroARepetir)
    elif numeroARepetir <0 or numeroARepetir >0:
        nombre = nombre + " "
        print( nombre * numeroARepetir)
        
programa() 
```

### Ejercicio 4
Crea un programa que dada una lista cualquiera, te diga si es simétrica o no. 
Si lo es, que te diga cuántos elementos tiene.


```python
frase = input().lower()
if str(frase) == "".join(reversed(frase)) :
    print("La frase es simetrica y tiene",len(frase), "elementos")
else:
    print("La frase no es simetrica")
```
