```python
import numpy as np
import pandas as pd

filename = 'C:/Users/Lucía Alvarez/OneDrive - psico.edu.uy/00 Barcelona activa/DelayedFlights.csv'

#Cargar los datos
data = pd.read_csv(filename) #, sep=',', encoding='utf8', index_col=0

#Mostrar los datos
print(data.head(3))


#Ver nombre de las columnas para eliminar algunas 
print("Los nombres de las columnas son", data.columns.values)
```

 Queda únicamente con las columnas que consideres relevantes.

```python
dataBorrado = data.drop(['Unnamed: 0', 'CRSDepTime',  'DepTime'
                         , 'CRSDepTime', 'ArrTime', 'CRSArrTime'
                         , 'FlightNum', 'TailNum'
                         , 'ActualElapsedTime', 'CRSElapsedTime'
                         , 'AirTime', 'DepDelay', 'CancellationCode'
                         , 'Diverted', 'CarrierDelay', 'NASDelay'
                         , 'SecurityDelay', 'LateAircraftDelay'
                         , 'TaxiIn', 'TaxiOut'
                         ,'Cancelled', 'WeatherDelay'], axis=1) #axis = 'columns', inplace=True

print("DataFrame despues de eliminar las columnas, los nombres de las columnas son:",dataBorrado.columns.values)
```

### Ejercicio 2
Realiza un informe completo del data frame:

Resume estadísticamente las columnas de interés
Encuentra cuántos datos faltantes hay por columna

```python
import numpy as np

class bcolors:
    Verde = '\033[92m' #GREEN
    Red = '\033[91m' #RED
    RESET = '\033[0m' #RESET COLOR

print(bcolors.Verde + "A continuación, se presenta una tabla que resume los estadísticos descriptivos de las variables seleccionadas: \n " + bcolors.RESET
     ,dataBorrado.describe())


print(bcolors.Verde + "En la tabla que sigue, se va a evaluar si las variables seleccionadas tienen valores perdidos: \n " + bcolors.RESET
     ,dataBorrado.isnull().any())

print(bcolors.Red + "ArrDelay es la única variable con valores perdidos del DataFrame" + bcolors.RESET)
```

Crear columna con vuelos atrasados

```python

print(bcolors.Red + "Un vuelo se considera puntual si ha operado en menos de 15 minutos."+ bcolors.RESET)

dataBorrado['Status'] = np.where(dataBorrado['ArrDelay']<=15, 'On Time', 'Late')

print("De los ", len(dataBorrado.ArrDelay), "vuelos realizados", len(dataBorrado[dataBorrado['Status'] == 'On Time']),
"estuvieron a tiempo y", len(dataBorrado[dataBorrado['Status'] == 'Late']), "arribaron 15 minutos más tarde de lo esperado")
```

Tabla de las aerolíneas con mayores retrasos acumulados

```python
Tabla = pd.crosstab(dataBorrado.UniqueCarrier, dataBorrado.Status)

print(bcolors.Verde + "A continuación, se presenta una tabla que idenfitica cuántos vuelos llegaron"
      "\n en el tiempo estipulado y los que llegaron tarde, según la aerolinea \n " + bcolors.RESET
     ,Tabla)
```
