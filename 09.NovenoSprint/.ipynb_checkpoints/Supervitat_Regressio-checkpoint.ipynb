```python
Regresion_1 = df.iloc[:, [15, 29]]

#Identificar valores perdidos
Regresion_1.isnull().sum()
```

```python
Regresion_1.dropna(subset = ["ArrDelay", "LateAircraftDelay"], axis = 0, inplace = True)

Regresion_1.isnull().sum()
```

```python
Regresion_1
```

```python
#Para evaluar si tienen la misma cantidad de respuestas

Regresion_1.describe()
```

```python
from sklearn.preprocessing import StandardScaler

#Estandarizacion de las variables
scaler = StandardScaler()
x_test_numericas = scaler.fit_transform(Regresion_1)

#transformación del array  con variables estandarizadas  en el dataframe
Regresion_1_limpia = pd.DataFrame(Regresion_1, columns=["ArrDelay", "LateAircraftDelay"])
Regresion_1_limpia
```

```python
fig = plt.figure(figsize =(10, 7)) 
  
ax = fig.add_axes([0, 0, 1, 1]) 
  
bp = ax.boxplot(Regresion_1_limpia) 
  
plt.show() 
```

```python
ax = sns.boxplot( y="ArrDelay", data=Regresion_1_limpia)
```

```python
ax = sns.boxplot( y="LateAircraftDelay", data=Regresion_1_limpia)
```

```python
import scipy.stats as stats

# encontrar Q1, Q3 y rango intercuartílico para cada columna
Q1 = Regresion_1.quantile (q = .25)
Q3 = Regresion_1.quantile (q = .75)
IQR = Regresion_1.apply (stats.iqr)

# solo mantenga las filas en el marco de datos que tengan valores dentro de 1.5 * IQR de Q1 y Q3
data_clean = Regresion_1 [~ ((Regresion_1 <(Q1-1.5 * IQR)) | (Regresion_1> (Q3 + 1.5 * IQR))). any (axis = 1)]

data_clean.isnull().sum()
```

```python
ax = sns.boxplot( y="ArrDelay", data=data_clean)
```

```python
fig = plt.figure(figsize =(10, 7)) 
  
ax = fig.add_axes([0, 0, 1, 1]) 
  
bp = ax.boxplot(data_clean) 
  
plt.show() 
```

```python
from sklearn.model_selection import train_test_split
# ArrDelay lo que queremos predecir segun LateA..
X = data_clean['LateAircraftDelay'].values.reshape(-1,1)
y = data_clean['ArrDelay'].values.reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() 
regressor.fit(X_train, y_train) #Entrena el algoritmo 
```

```python
#Para obtener el intercepto:
print(regressor.intercept_)

#Para obtener la pendiente
print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test)

df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df.head(25)
```

```python
df1 = df.head(25)
df1.plot(kind='bar',figsize=(10,8))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```
