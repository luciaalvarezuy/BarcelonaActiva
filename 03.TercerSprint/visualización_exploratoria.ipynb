```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

filename = 'C:/Users/Lucía Alvarez/OneDrive - psico.edu.uy/00 Barcelona activa/DelayedFlights.csv'



#Cargar los datos
DataFlights = pd.read_csv(filename) 
```

##  Una variable categórica (UniqueCarrier)

```python
titulo = "Cantidad de vuelos por operador"
plt.figure(figsize =(9, 9))
fig1 = sns.countplot(y="UniqueCarrier", data=DataFlights, order=DataFlights['UniqueCarrier'].value_counts().index)
plt.yticks(rotation=10)
plt.title(titulo, fontsize=20, color='#6146C6', alpha=0.8)
plt.ylabel('Operadores')
plt.xlabel('Cantidad de vuelos')
plt.show()

fig1.figure.savefig("fig1.png")
```

##  Una variable numérica (ArrDelay)

```python
titulo = "Gráfico de ArrDelay"
plt.figure(figsize =(9, 9))
fig2 = sns.kdeplot(DataFlights['ArrDelay'], shade=True)
plt.title(titulo, fontsize=20, color='#6146C6', alpha=0.8)
plt.show()

fig2.figure.savefig('fig2.png')
```

##  Una variable numérica y una categórica (ArrDelay y  UniqueCarrier)

```python
titulo = "Gráfico de ArrDelay y UniqueCarrier"
sns.set_style("whitegrid")
fig3 = sns.catplot(x="UniqueCarrier", y="ArrDelay", data=DataFlights, height=5, aspect=2)
plt.title(titulo, fontsize=20, color='#6146C6', alpha=0.8)
plt.xlabel('Unique carrier code')
plt.show()

fig3.figure.savefig('fig3.png')
```

## Dos variables numéricas (ArrDelay y DepDelay)


```python
data = DataFlights[["ArrDelay", "DepDelay"]]
plt.figure(figsize =(50, 30))
fig4 = sns.relplot(data=data, x="ArrDelay", y="DepDelay", kind="scatter")
plt.show()
fig4.figure.savefig('fig4.png')
```

## Tres variables (ArrDelay, DepDelay y UniqueCarrier)


```python
fig5= sns.lmplot( x="ArrDelay", y="DepDelay", data=DataFlights, fit_reg=False, hue='UniqueCarrier', legend=False)
plt.legend(loc='lower right')
plt.show()

fig5.figure.savefig('fig5.png')
```

## Más de tres variables (ArrDelay, DepDelay, AirTime y UniqueCarrier).
 ArrDelay: numerica
DepDelay: numerica
AirTime: numerica
UniqueCarrier: categorica


```python
fig6 = sns.scatterplot(data = DataFlights, x = "ArrDelay", y = "DepDelay", hue = "UniqueCarrier", size = "AirTime")
plt.legend(loc='best')
plt.legend(bbox_to_anchor=(1, 1), loc=2) 
plt.show()
fig6.figure.savefig('fig6.png', dpi=300, bbox_inches='tight')
```
